```python
# 图像api 分为如下几大类：
# resize 改变大小,缩放
# crop 裁剪图像
# flip 翻转图像
# brightness & contrast 改变光照和对比度

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
```

```python
name = './gugong.jpg'
img_string = tf.read_file(name) # 首先读入图片为一个字符串
img_decoded = tf.image.decode_image(img_string)

sess = tf.Session()
img_decoded_val = sess.run(img_decoded)
print(img_decoded_val.shape)

%matplotlib inline
imshow(img_decoded_val)
```

```python
# resize 
# tf.image.resize_area
# tf.image.resize_bicubic 当图片缩小时没有损失，当图像被放大时使用
# tf.image.resize_nearest_neighbor 使用最邻近的像素点进行放大时赋值填充

name = './gugong.jpg'
img_string = tf.read_file(name) # 首先读入图片为一个字符串
img_decoded = tf.image.decode_image(img_string)

#将其变成思维向量，第一维为1
img_decoded = tf.reshape(img_decoded, [1, 318,620,3])

resize_img = tf.image.resize_bicubic(img_decoded, [626, 1240]) # 将长宽都加倍

sess = tf.Session()
img_decoded_val = sess.run(resize_img)

# 再将输出转换为三维的
img_decoded_val = img_decoded_val.reshape((626, 1240, 3))
# 将结果转换为int类型，原来是float
img_decoded_val = np.asarray(img_decoded_val, np.uint8)
print(img_decoded_val.shape)

%matplotlib inline
imshow(img_decoded_val)
```

```python
# crop
# tf.image.pad_to_bounding_box 将图像放在画布上，这个画布可以用pad填充，图像的周边都是黑色的
# tf.image.crop_to_bounding_box 跟上面的区别是，上面是填充，它是裁剪
# tf.random_crop

name = './gugong.jpg'
img_string = tf.read_file(name) # 首先读入图片为一个字符串
img_decoded = tf.image.decode_image(img_string)

#将其变成思维向量，第一维为1
img_decoded = tf.reshape(img_decoded, [1, 318,620,3])

padded_img = tf.image.pad_to_bounding_box(img_decoded, 
                                         50, 100, 500, 800) # 在50,100的位置上，裁剪500 800 

sess = tf.Session()
img_decoded_val = sess.run(padded_img)

# 再将输出转换为三维的
img_decoded_val = img_decoded_val.reshape((500, 800, 3))
# 将结果转换为int类型，原来是float
img_decoded_val = np.asarray(img_decoded_val, np.uint8)
print(img_decoded_val.shape)

%matplotlib inline
imshow(img_decoded_val)


```

```python
# flip
# tf.image.flip_up_down
# tf.image.flip_left_right
# tf.image.random_flip_up_down
# tf.image.random_flip_left_right

name = './gugong.jpg'
img_string = tf.read_file(name) # 首先读入图片为一个字符串
img_decoded = tf.image.decode_image(img_string)

#将其变成思维向量，第一维为1
img_decoded = tf.reshape(img_decoded, [1, 318,620,3])

flipped_img = tf.image.flip_up_down(img_decoded)

sess = tf.Session()
img_decoded_val = sess.run(flipped_img)

# 再将输出转换为三维的
img_decoded_val = img_decoded_val.reshape((318, 620, 3))
# 将结果转换为int类型，原来是float
img_decoded_val = np.asarray(img_decoded_val, np.uint8)
print(img_decoded_val.shape)

%matplotlib inline
imshow(img_decoded_val)
```

```python
# brightness & contrast 改变光照和对比度: 对一张普通人来生成一张在普通光照下的人脸数据和在光照增强，减弱下的人脸数据，用来扩展数据用的！
# tf.image.adjust_brightness
# tf.image.random_brightness
# tf.image.adjust_constrast
# tf.image.random_constrast

name = './gugong.jpg'
img_string = tf.read_file(name) # 首先读入图片为一个字符串
img_decoded = tf.image.decode_image(img_string)

#将其变成思维向量，第一维为1
img_decoded = tf.reshape(img_decoded, [1, 318,620,3])

brightness_img = tf.image.adjust_brightness(img_decoded, 0.5)

sess = tf.Session()
img_decoded_val = sess.run(brightness_img)

# 再将输出转换为三维的
img_decoded_val = img_decoded_val.reshape((318, 620, 3))
# 将结果转换为int类型，原来是float
img_decoded_val = np.asarray(img_decoded_val, np.uint8)
print(img_decoded_val.shape)

%matplotlib inline
imshow(img_decoded_val)

```
