# matplotlib基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100) # 等分0~10，等分100段
x
```

```python
y = np.sin(x)
y
```

```python
z = np.cos(x)
z
```

```python
# 开始使用plt, 绘制一个以x为横轴，y为纵轴的图形  正弦
plt.plot(x, y)
```

```python
# plt.show() # 貌似这个版本不需要调用show()就可以显示了
```

```python
plt.plot(x, z)  # 余弦图
```

```python
z.shape
```

```python
plt.plot(x, y)
plt.plot(x, z)
plt.show()
```

```python
plt.plot(x, y)
plt.plot(x, z, color="red") # 可以使用16进制码
plt.show()
```

```python
plt.plot(x, y)
plt.plot(x, z, color="red", linestyle="--") # 可以使用16进制码
plt.show()
```

```python
plt.plot(x, y)
plt.plot(x, z, color="red", linestyle="--") # 可以使用16进制码
plt.xlim(-5, 15)
plt.show()
```

```python
plt.plot(x, y)
plt.plot(x, z, color="red", linestyle="--") # 可以使用16进制码
plt.xlim(-5, 15)
plt.ylim(0, 1.5)
plt.show()
```

```python
plt.plot(x, y)
plt.plot(x, z, color="red", linestyle="--") # 可以使用16进制码
plt.axis([-1,11, -2, 2]) # 还可以一起调节x,y显示幅度
plt.show()
```

```python
plt.plot(x, y)
plt.plot(x, z, color="red", linestyle="--") # 可以使用16进制码
plt.axis([-1,11, -2, 2]) # 还可以一起调节x,y显示幅度
plt.xlabel("x axis")
plt.ylabel("y value")
plt.show()
```

```python
plt.plot(x, y, label = "sin(x)")
plt.plot(x, z, color="red", linestyle="--", label = "cos(x)") # 可以使用16进制码
plt.axis([-1,11, -2, 2]) # 还可以一起调节x,y显示幅度
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.show()
```

```python
plt.plot(x, y, label = "sin(x)")
plt.plot(x, z, color="red", linestyle="--", label = "cos(x)") # 可以使用16进制码
plt.axis([-1,11, -2, 2]) # 还可以一起调节x,y显示幅度
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.title("Welcome to the ML world!")
plt.show()
```

### 散点图 Scatter Plot

```python
# 散点图通常用来在横纵坐标上都表示特征

plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x, y)
plt.scatter(x, z)
plt.show()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
x
```

```python
plt.scatter(x, y, alpha=0.5) # alpha透明度
plt.show()
```
