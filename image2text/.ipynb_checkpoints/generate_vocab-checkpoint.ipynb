```python
import sys
import os
import pprint

input_descrption_file = './data/results.token'
output_vocab_file = './data/vocab.txt'

def generate_vocab(input_file):
    with open(input_file, 'r') as f:
        lines = f.readlines()
    max_length_of_sentence = 0
    length_dict = {}
    vocab_dict = {}
    for line in lines:
        file_name, descrption = line.strip('\n').split('\t')
        words = descrption.strip(' ').split() # 英文直接按空格分割即可
        max_length_of_sentence = max(max_length_of_sentence, len(descrption))
        length_dict.setdefault(len(words), 0)
        length_dict[len(words)] += 1
        
        for word in words:
            vocab_dict.setdefault(word, 0)
            vocab_dict[word] += 1
    print(max_length_of_sentence)
    pprint.pprint(length_dict) #可以用一个格式比较好的方式打印list, dict等
    return vocab_dict
        
generate_vocab(input_descrption_file)
```
